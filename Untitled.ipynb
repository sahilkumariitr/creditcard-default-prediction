```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('train_dataset_final1.csv')
validate_data = pd.read_csv('validate_dataset_final.csv')
```

### Data Preprocessing :-

```python
train_data = train_data.drop('Customer_ID', axis = 1)
# since it wont be useful in modeling
```

```python
train_data.columns
```

```python
#since from info we got that the age coloumn has missing values so lets just fill it with median
train_data['age'] = train_data['age'].fillna(train_data['age'].median())
```

```python
train_data.info()
```

```python
#now we will remove the target column next_month_default from training features and store it in a separate variable
target = train_data['next_month_default']
train_data = train_data.drop(columns=['next_month_default'])
```

```python
train_data.columns
```

```python
# since customer id wont be useful in modelling
train_data = train_data.drop('Customer_ID', axis = 1)
```

```python
validate_data = validate_data.drop('Customer_ID', axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the training features
train_scaled = pd.DataFrame(scaler.fit_transform(train_data), columns=train_data.columns)

# Apply the same transformation to the validation set
validate_scaled = pd.DataFrame(scaler.transform(validate_data), columns=validate_data.columns)
```

### EDA :-

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Plot
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='next_month_default', hue='next_month_default', data=train_data, palette='crest', legend=False)

# Annotate with percentages
total = len(train_data)
for p in ax.patches:
    height = p.get_height()
    percent = 100 * height / total
    ax.annotate(f'{percent:.2f}%', 
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='bottom', fontsize=11)

# Fix tick labels safely
ax.set_xticks([0, 1])
ax.set_xticklabels(['No Default (0)', 'Default (1)'])

# Titles and labels
ax.set_title('Target Variable Distribution: next_month_default')
ax.set_xlabel('Default Status')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python

# List of key features
key_features = ['LIMIT_BAL', 'age', 'AVG_Bill_amt', 'PAY_TO_BILL_ratio']

# Summary statistics table
summary_stats = train_data[key_features].describe()
print(summary_stats)

# Optional: Visualize distributions
plt.figure(figsize=(12, 8))
for i, col in enumerate(key_features):
    plt.subplot(2, 2, i + 1)
    sns.histplot(train_data[col], kde=True, bins=50, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Create a new dataframe with the target included
data_with_target = train_data.copy()

# Compute correlation matrix
corr_matrix = data_with_target.corr()

# Set up the matplotlib figure
plt.figure(figsize=(14, 12))

# Generate heatmap
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": .75})

# Title
plt.title('Correlation Heatmap of Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
corr_with_target = corr_matrix['next_month_default'].abs().sort_values(ascending=False)
print(corr_with_target)
```

```python
# Group data by marital status and compute mean default rate
marriage_default = train_data.groupby('marriage')['next_month_default'].mean().reset_index()

# Map numeric codes to readable labels
marriage_default['marriage'] = marriage_default['marriage'].map({
    1: 'Married',
    2: 'Single',
    3: 'Others'
})

# Plot
plt.figure(figsize=(6, 4))
sns.barplot(data=marriage_default, x='marriage', y='next_month_default', hue='marriage', legend=False, palette='Set2')
plt.title('Default Rate by Marital Status')
plt.ylabel('Default Rate')
plt.xlabel('Marital Status')
plt.ylim(0, 0.4)
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, classification_report

# Assuming:
# train_data = training features (without target)
# target = training target (e.g., next_month_default)
# valid_data = validation features (without target)
# valid_target = validation target

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train_data)
X_valid_scaled = scaler.transform(validate_data)

# Train the model
model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(X_train_scaled, target)

# Predict on validation set
y_pred = model.predict(X_valid_scaled)
y_proba = model.predict_proba(X_valid_scaled)[:, 1]

# Evaluate
print("Validation Accuracy:", accuracy_score(valid_target, y_pred))
print("Validation ROC-AUC:", roc_auc_score(valid_target, y_proba))
print("Confusion Matrix:\n", confusion_matrix(valid_target, y_pred))
print("Classification Report:\n", classification_report(valid_target, y_pred))
```
